```python
from langchain_openai import OpenAIEmbeddings
from langchain_chroma  import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books=pd.read_csv("books_cleaned.csv")
```

```python
books["tagged_description"].to_csv("tagged_description.txt",

                                   index=False,
                                   header=False
                                   )
```

```python
raw_data=TextLoader("tagged_description.txt",encoding="utf-8").load()
text_splitter=CharacterTextSplitter(chunk_size= 10_000 ,chunk_overlap = 0,separator="\n")
document=text_splitter.split_documents(raw_data)
```

```python
document[0]
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embedder=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
db_books= Chroma.from_documents(document,embedding=embedder)
```
